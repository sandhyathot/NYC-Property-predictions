## Modelling


### Data Preparation

```python
columns=['borough','building_class_category','commercial_units','gross_square_feet',
         'sale_price','building_age','land_square_feet','residential_units']
data_model=df.loc[:,columns]
```

```python
one_hot_featu=['borough','building_class_category']
longest_str=max(one_hot_featu,key=len)
total_num_uni_cat=0
for feature in one_hot_featu:
    num_unique=len(df[feature].unique())
    print('{col:<{fill_col}}:{num:d} unique categorical values.'.format(col=feature,fill_col=len(longest_str),num=num_unique))
    total_num_uni_cat+=num_unique
print('{total: d} columns will be added during one hot encoding.'.format(total=total_num_uni_cat))
```

```python
one_hot_encoded=pd.get_dummies(data_model[one_hot_featu])
one_hot_encoded.info(verbose=True,memory_usage=True, null_counts=True)
```

```python
data_model=data_model.drop(one_hot_featu,axis=1)
```

```python
data_model=pd.concat([data_model,one_hot_encoded],axis=1)
data_model.head()
```

```python
plt.figure(figsize=(10,6))
sns.distplot(data_model['sale_price'])
plt.title('Histogram of Sale price')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
data_model['sale_price']=StandardScaler().fit_transform(np.log(data_model['sale_price']).values.reshape(-1,1))
```

```python
plt.figure(figsize=(10,6))
sns.distplot(data_model['sale_price'])
plt.title('Histogram of Normalised Sales Price')
plt.show()
```

```python
data_model.describe()
```

```python
data_model['commercial_units']=data_model['commercial_units']+1
data_model['residential_units']=data_model['residential_units']+1
data_model['commercial_units']=StandardScaler().fit_transform(np.log(data_model['commercial_units']).values.reshape(-1,1))
data_model['residential_units']=StandardScaler().fit_transform(np.log(data_model['residential_units']).values.reshape(-1,1))
```

```python
data_model['gross_square_feet']=data_model['gross_square_feet']+1
data_model['land_square_feet']=data_model['land_square_feet']+1
data_model['gross_square_feet']=StandardScaler().fit_transform(np.log(data_model['gross_square_feet']).values.reshape(-1,1))
data_model['land_square_feet']=StandardScaler().fit_transform(np.log(data_model['land_square_feet']).values.reshape(-1,1))
```

```python
data_model['building_age']=data_model['building_age']+1
data_model['building_age']=StandardScaler().fit_transform(np.log(data_model['building_age']).values.reshape(-1,1))
```

```python
plt.figure(figsize=(10,6))
sns.distplot(data_model['building_age'])
plt.title('Normalized building age')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.distplot(data_model['commercial_units'])
plt.title('Normalized commercial units')
plt.show()
```

```python
x=data_model.drop(['sale_price'],axis=1)
```

```python
y=data_model ['sale_price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
```

```python
cv_score_linreg=cross_val_score(reg,x_train,y_train,cv=5)
```

```python
print('R^2: {}'.format(reg.score(x_test,y_test)))
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print('root mean squared error: {}'.format(rmse))
print('Average 5 fold cv score: {}'.format(np.mean(cv_score_linreg)))
print(cv_score_linreg)
```

```python
rf_reg=RandomForestRegressor()
rf_reg.fit(x_train,y_train)
y_predict_rfreg=rf_reg.predict(x_test)
cv_score_rfreg=cross_val_score(rf_reg,x_train,y_train,cv=5)

print('R^2: {}'.format(rf_reg.score(x_test,y_test)))
rmse=np.sqrt(mean_squared_error(y_test,y_predict_rfreg))
print('root mean squared error: {}'.format(rmse))
print('Average 5 fold cv score: {}'.format(np.mean(cv_score_rfreg)))
print(cv_score_rfreg)
```

## Feature Importance of Random Forest

```python
importance=pd.DataFrame(list(zip(x_train.columns,np.transpose(rf_reg.feature_importances_)))).sort_values(1,ascending=False)
importance
```

```python
importances=rf_reg.feature_importances_
std=np.std([tree.feature_importances_ for tree in rf_reg.estimators_],axis=0)
indices=np.argsort(importances)[::-1]

#Plot the feature imporance of the forest
plt.figure()
plt.title('Feature Importances')
plt.bar(range(x_train.shape[1]),importances[indices],color='r',yerr=std[indices],align='center')
plt.xticks(range(x_train.shape[1]),x_train.columns[indices],rotation=90)
plt.xlim([-1,10])
plt.show()
```

```python
ridge_reg=Ridge()
ridge_reg.fit(x_train,y_train)
y_pred_ridge=ridge_reg.predict(x_test)
```

```python
cv_score_ridge=cross_val_score(ridge_reg,x_train,y_train,cv=5)

print('R^2: {}'.format(ridge_reg.score(x_test,y_test)))
rmse=np.sqrt(mean_squared_error(y_test,y_pred_ridge))
print('root mean squared error: {}'.format(rmse))
print('average cv score: {}'.format(np.mean(cv_score_ridge)))
print(cv_score_ridge)
```

```python
l1_space=np.linspace(0,1,30)
param_grid={'l1_ratio':l1_space}
```

```python
elastic_net=ElasticNet()
gm_cv=GridSearchCV(elastic_net,param_grid,cv=5)
gm_cv.fit(x_train,y_train)
y_pred_elas=gm_cv.predict(x_test)
r2=gm_cv.score(x_test,y_test)
mse=mean_squared_error(y_test,y_pred_elas)
```

```python
print('Tuned Elastic l1 ration: {}'.format(gm_cv.best_params_))
print('Tuned Elastic R squared: {}'.format(r2))
print('Tuned ElasticNet MSE: {}'.format(mse))
```
